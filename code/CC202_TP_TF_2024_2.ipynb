```python
from pyvis.network import Network
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
```

```python
def draw_graph(G):
    net = Network(notebook=True, directed=True)

    cmap = plt.get_cmap("viridis")

    max_in_degree = max(dict(G.in_degree).values())
    min_in_degree = min(dict(G.in_degree).values())

    for node in G.nodes:
        in_degree = G.in_degree(node)

        normalized_in_degree = (in_degree - min_in_degree) / (max_in_degree - min_in_degree) if max_in_degree != min_in_degree else 0.5
        color = mcolors.to_hex(cmap(normalized_in_degree))
        net.add_node(node, size=10 + in_degree * 2.5, color=color)

    for edge in G.edges:
        net.add_edge(edge[0], edge[1], arrows='to')
    net.set_options("""
    var options = {
    "physics": {
        "barnesHut": {
        "gravitationalConstant": -30000,
        "springLength": 200,
        "springStrength": 100
        },
        "repulsion": {
        "centralGravity": 0.1,
        "springLength": 200,
        "springStrength": 100,
        "damping": 0.9
        },
        "minVelocity": 0.75
    }
    }
    """)

    net.show("grafo.html")
```

```python
draw_graph(G)
```
